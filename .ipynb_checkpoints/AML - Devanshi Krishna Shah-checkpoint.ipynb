### The dataframe below proves that: 
- All fraud transactions are from accounts flagged as fraud
- All accounts involved in any fraud transaction (1 or many transactions) are flagged as fraud
- All transactions involving an account flagged as fraud are not fraudulent i.e. accounts flagged fraud are a part of non fraudulent transactions
- There do exist cases where accounts not involved in any fraudulent transaction as a sender or a receiver are flagged fraud

#### The conclusion drawn here is it would be preferable to use unsupervised ML to flag accounts as fraud because supervised learning can lead to several False-Positives as even involvement in 1 fraud transaction or 0 fraud transaction can mark an account fraud which in real world cases is not suitable owing to factors like an account can be hacked for that transaction etc. 

```python
# List of ACCOUNT_ID values flagged as fraudulent
fraudulent_account_ids = accounts_df[accounts_df['IS_FRAUD'] == True]['ACCOUNT_ID'].tolist()

# Filter transactions where either SENDER_ACCOUNT_ID or RECEIVER_ACCOUNT_ID is in the list of fraudulent account IDs
transactions_sender = txs_df[(txs_df['SENDER_ACCOUNT_ID'].isin(fraudulent_account_ids))]
transactions_receiver = txs_df[(txs_df['RECEIVER_ACCOUNT_ID'].isin(fraudulent_account_ids))]

# How many fraudulent transactions are these account_ids a part of as sender or receiver
fraudulent_sender_transaction_counts = transactions_sender[transactions_sender['IS_FRAUD']==True]['SENDER_ACCOUNT_ID'].value_counts().reset_index()
fraudulent_sender_transaction_counts.columns = ['ACCOUNT_ID', 'FRAUDULENT_SENDER_COUNT']
fraudulent_receiver_transaction_counts = transactions_receiver[transactions_receiver['IS_FRAUD']==True]['RECEIVER_ACCOUNT_ID'].value_counts().reset_index()
fraudulent_receiver_transaction_counts.columns = ['ACCOUNT_ID', 'FRAUDULENT_RECEIVER_COUNT']

# Merge fraudulent sender and receiver counts
fraudulent_transaction_counts = pd.merge(fraudulent_sender_transaction_counts, fraudulent_receiver_transaction_counts, on='ACCOUNT_ID', how='outer').fillna(0)


# How many fraudulent transactions are these account_ids a part of as sender or receiver
non_fraudulent_sender_transaction_counts = transactions_sender[transactions_sender['IS_FRAUD']==False]['SENDER_ACCOUNT_ID'].value_counts().reset_index()
non_fraudulent_sender_transaction_counts.columns = ['ACCOUNT_ID', 'NON_FRAUDULENT_SENDER_COUNT']
non_fraudulent_receiver_transaction_counts = transactions_receiver[transactions_receiver['IS_FRAUD']==False]['RECEIVER_ACCOUNT_ID'].value_counts().reset_index()
non_fraudulent_receiver_transaction_counts.columns = ['ACCOUNT_ID', 'NON_FRAUDULENT_RECEIVER_COUNT']

# Merge fraudulent sender and receiver counts
non_fraudulent_transaction_counts = pd.merge(non_fraudulent_sender_transaction_counts, non_fraudulent_receiver_transaction_counts, on='ACCOUNT_ID', how='outer').fillna(0)

# Now, merge fraudulent and non-fraudulent transaction counts
combined_transaction_counts = pd.merge(fraudulent_transaction_counts, non_fraudulent_transaction_counts, on='ACCOUNT_ID', how='outer').fillna(0)

# Convert the columns to integer type
combined_transaction_counts['FRAUDULENT_SENDER_COUNT'] = combined_transaction_counts['FRAUDULENT_SENDER_COUNT'].astype(int)
combined_transaction_counts['FRAUDULENT_RECEIVER_COUNT'] = combined_transaction_counts['FRAUDULENT_RECEIVER_COUNT'].astype(int)
combined_transaction_counts['NON_FRAUDULENT_SENDER_COUNT'] = combined_transaction_counts['NON_FRAUDULENT_SENDER_COUNT'].astype(int)
combined_transaction_counts['NON_FRAUDULENT_RECEIVER_COUNT'] = combined_transaction_counts['NON_FRAUDULENT_RECEIVER_COUNT'].astype(int)


combined_transaction_counts
```

- removing the outliers from txs_df 
    - there are only 5 transactions for amounts ranging from 0 to 5 for non fraudulent transactions, however I'm keeping them as over 70 of fraudulent transactions are in the range of 2.5 to 5. Hence, keeping non fraudulent transactions in this range too might help the model learn patterns
    - for non fraudulent transactions the outliers are in the range of 9000000 to 12000000, since there are only 3+3+1 = 7 transaction, I believe they can be removed from consideration
- combining the dataframes with all the required columns 
- adding columns using sender_account_id and receiver_account_id so that I can drop them during modelling 
    - extracting more information using sender_account_id and receiver_account_id such as frequency and recency which is the number of transactions or the time since the last transaction for each sender account as a sender and each receiver account as a receiver

- standardizing the tx_amount column 
    - it is necessary too as it has a wide range and to give each of them equal wieght in the model. I'm preferring standardization over normalization as standardization is comparitively less senssitive to outliers and for normalization choosing the extreme min max values might not be ideal as very few transactions lie in the extremes
- correlation
- sampling for dealing with imbalance in train dataset

## ML Modelling
### Helper functions

```python
def conf_matrix_percent(conf_matrix_train, conf_matrix_val, conf_matrix_test, model, label):
    fig, axs = plt.subplots(1, 3, figsize=(30, 10))

    LABELS = ["Normal", "Fraud"]
    titles = ["Train", "Validation", "Test"]
    conf_matrices = [conf_matrix_train, conf_matrix_val, conf_matrix_test]

    for idx, conf_matrix in enumerate(conf_matrices):
        sns.heatmap(conf_matrix / np.sum(conf_matrix), annot=True, fmt='.3%', cmap='Blues',
                    xticklabels=LABELS, yticklabels=LABELS, annot_kws={"size": 20}, ax=axs[idx])
        axs[idx].set_title(titles[idx], fontsize=25)
        axs[idx].tick_params(axis='both', labelsize=18)  # Adjust tick label size

    plt.suptitle("{} Confusion Matrix for {}".format(model,label), fontsize=25)
    plt.subplots_adjust(wspace=2)
    plt.tight_layout()

    # Adjust the position of axis labels to avoid overlapping with heatmaps
    fig.text(0.5, -0.1, 'Predicted class', ha='center', fontsize=28)
    fig.text(-0.01, 0.5, 'True class', va='center', rotation='vertical', fontsize=28)
    plt.show()
```

```python
def evaluation(y_train,y_train_pred,y_val_pred,y_val,y_test, y_test_pred, model,label):
    cm_train = confusion_matrix(y_train, y_train_pred)
    cm_val = confusion_matrix(y_val, y_val_pred)
    cm_test = confusion_matrix(y_test, y_test_pred)
    conf_matrix_percent(cm_train,cm_val,cm_test, model, label)

    report_train = classification_report(y_train, y_train_pred, zero_division=0)
    report_val = classification_report(y_val, y_val_pred, zero_division=0)
    report_test = classification_report(y_test, y_test_pred, zero_division=0)


    print(f"\n{'Classification Report for Validation ' + label.ljust(23)}{'Classification Report for Test ' + label}")
    for line_val, line_test in zip(report_val.split('\n'), report_test.split('\n')):
        print(f"{line_val.ljust(60)}{line_test}")
    
    print("\nConfusion Matrix for Train".ljust(40), f"{'Confusion Matrix for Validation'.ljust(40)}{'Confusion Matrix for Test'}")
    tn_train, fp_train, fn_train, tp_train = cm_train.ravel()
    tn_val, fp_val, fn_val, tp_val = cm_val.ravel()
    tn_test, fp_test, fn_test, tp_test = cm_test.ravel()

    print(f"tn = {tn_train} \t fp = {fp_train},".ljust(35), f"tn = {tn_val} \t fp = {fp_val},".ljust(35), f"tn = {tn_test} \t fp = {fp_test},")
    print(f"fn = {fn_train} \t tp = {tp_train}".ljust(35), f"fn = {fn_val} \t tp = {tp_val}".ljust(35), f"fn = {fn_test} \t tp = {tp_test}")
```

```python
def pr_auc(classifier,X,y):
    y_proba = classifier.predict_proba(X)[:, 1]
    precision, recall, thresholds = precision_recall_curve(y, y_proba)


    # Initialize variables to store the optimal threshold and the highest F1 score
    optimal_threshold = 0.0
    optimal_precision = 0.0
    optimal_recall = 0.0
    max_f1_score = 0.0

    # Find the threshold with the highest F1 score
    for i in range(len(precision)):
        pr = round(precision[i],2)
        re = round(recall[i],2)
        f1 = round(((2 * pr * re) / (pr + re + 1e-10)),2)  # Avoid division by zero
        if f1 > max_f1_score:
            max_f1_score = f1
            optimal_threshold = round(thresholds[i],2)
            optimal_precision = pr
            optimal_recall = re

    print(f"Optimal Threshold using F1 Score: {optimal_threshold}")
    print(f"Optimal Precision using F1 Score: {optimal_precision}")
    print(f"Optimal Recall using F1 Score: {optimal_recall}")

    # Plot Precision-Recall curve
    plt.figure(figsize=(6, 4))
    plt.plot(recall, precision, color='blue', lw=2, label='Precision-Recall curve')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend(loc="lower left")
    plt.show()
    return round(optimal_threshold,2)
```

```python
def modelling(classifier, X_train, y_train, X_val, y_val, X_test, y_test,model,itr_no,label):
    classifier.fit(X_train,y_train)
    threshold = pr_auc(classifier, X_val, y_val)
    
    
    y_train_pred_proba = classifier.predict_proba(X_train)[:, 1]
    y_val_pred_prob = classifier.predict_proba(X_val)[:, 1]
    y_test_pred_proba = classifier.predict_proba(X_test)[:, 1]
   
    # Update predictions based on the new threshold
    y_train_pred = (y_train_pred_proba >= threshold).astype(int)
    y_val_pred = (y_val_pred_prob >= threshold).astype(int)
    y_test_pred = (y_test_pred_proba >= threshold).astype(int)
    
    evaluation(y_train,y_train_pred,y_val,y_val_pred,y_test,y_test_pred,f"{model} Iteration {itr_no}",label)
```

### Multi - Label SMOTE (that could be used if our data was not time-series)

```python
def get_tail_label(df):
    """
    Give tail label colums of the given target dataframe
    
    args
    df: pandas.DataFrame, target label df whose tail label has to identified
    
    return
    tail_label: list, a list containing column name of all the tail label
    """
    columns = df.columns
    n = len(columns)
    irpl = np.zeros(n)
    for column in range(n):
        irpl[column] = df[columns[column]].value_counts()[1]
    irpl = max(irpl)/irpl
    mir = np.average(irpl)
    tail_label = []
    for i in range(n):
        if irpl[i] > mir:
            tail_label.append(columns[i])
    return tail_label

def get_index(df):
    """
    give the index of all tail_label rows
    args
    df: pandas.DataFrame, target label df from which index for tail label has to identified

    return
    index: list, a list containing index number of all the tail label
    """
    tail_labels = get_tail_label(df)
    index = set()
    for tail_label in tail_labels:
        sub_index = set(df[df[tail_label]==1].index)
        index = index.union(sub_index)
    return list(index)

def get_minority_instace(X, y):
    """
    Give minority dataframe containing all the tail labels
    
    args
    X: pandas.DataFrame, the feature vector dataframe
    y: pandas.DataFrame, the target vector dataframe
    
    return
    X_sub: pandas.DataFrame, the feature vector minority dataframe
    y_sub: pandas.DataFrame, the target vector minority dataframe
    """
    index = get_index(y)
    X_sub = X[X.index.isin(index)].reset_index(drop = True)
    y_sub = y[y.index.isin(index)].reset_index(drop = True)
    return X_sub, y_sub

def nearest_neighbour(X):
    """
    Give index of 5 nearest neighbor of all the instance
    
    args
    X: np.array, array whose nearest neighbor has to find
    
    return
    indices: list of list, index of 5 NN of each element in X
    """
    nbs=NearestNeighbors(n_neighbors=5,metric='euclidean',algorithm='kd_tree').fit(X)
    euclidean,indices= nbs.kneighbors(X)
    return indices

def MLSMOTE(X,y, n_sample):
    """
    Give the augmented data using MLSMOTE algorithm
    
    args
    X: pandas.DataFrame, input vector DataFrame
    y: pandas.DataFrame, feature vector dataframe
    n_sample: int, number of newly generated sample
    
    return
    new_X: pandas.DataFrame, augmented feature vector data
    target: pandas.DataFrame, augmented target vector data
    """
    indices2 = nearest_neighbour(X)
    n = len(indices2)
    new_X = np.zeros((n_sample, X.shape[1]))
    target = np.zeros((n_sample, y.shape[1]))
    for i in range(n_sample):
        reference = random.randint(0,n-1)
        neighbour = random.choice(indices2[reference,1:])
        all_point = indices2[reference]
        nn_df = y[y.index.isin(all_point)]
        ser = nn_df.sum(axis = 0, skipna = True)
        target[i] = np.array([1 if val>2 else 0 for val in ser])
        ratio = random.random()
        # Calculate the absolute difference to ensure positivity
        gap = np.abs(X.loc[reference, :] - X.loc[neighbour, :])
        new_X[i] = np.array(X.loc[reference,:] + ratio * gap)
    new_X = pd.DataFrame(new_X, columns=X.columns)
    
    # Round the specified float columns to integers
    columns_to_round = ['SENDER_ACCOUNT_ID','RECEIVER_ACCOUNT_ID','TIMESTAMP', 'SENDER_TIME_DIFF', 'SENDER_FREQUENCY', 'RECEIVER_TIME_DIFF', 'RECEIVER_FREQUENCY']
    new_X[columns_to_round] = new_X[columns_to_round].round().astype(int)
    
    # Create empty columns for 'SENDER_ACCOUNT_ID' and 'RECEIVER_ACCOUNT_ID'
    new_X['SENDER_ACCOUNT_ID'] = np.nan
    new_X['RECEIVER_ACCOUNT_ID'] = np.nan
    
    # Fill NaN values in 'SENDER_ACCOUNT_ID' and 'RECEIVER_ACCOUNT_ID' with -1 (or any other suitable integer)
    new_X['SENDER_ACCOUNT_ID'].fillna(-1, inplace=True)
    new_X['RECEIVER_ACCOUNT_ID'].fillna(-1, inplace=True)

    new_X = new_X.astype({'SENDER_ACCOUNT_ID': int, 'RECEIVER_ACCOUNT_ID': int})

    target = pd.DataFrame(target, columns=y.columns)
    return new_X, target
```
